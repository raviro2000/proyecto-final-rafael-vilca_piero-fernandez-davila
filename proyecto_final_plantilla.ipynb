## Ejercicio con data wrangling:
1) Relevancia:
Para esta sección se recoge información del reporte mundial de felicidad. La literatura sobre la medición de la felicicad retrocede a los intentos de medición de probreza multidimencional. Una vez se reconoce la multidimensionalidad del bienestar y la pobreza queda más claro que el PBI pc no es el unico elemento importante en la determinación del desarrollo en los países. En este caso, se aplicara un analisis de indicadores de bienestar autopercibido (percepción de corrupción, esperanza de vida al nacer, satisfacción-life ladder) en comparación con el PBI pc de los países miembros de la Alianza del Pacífico.


```python
#procedemos a ingresar datos de la felicidad
import pandas as pd
# URL del conjunto de datos
url = "https://raw.githubusercontent.com/ccsuehara/python_para_las_ccss/main/Clase%205/data/world-happiness-report-2023.csv"

# Cargar el conjunto de datos en un DataFrame
felicidad = pd.read_csv(url)
felicidad
```

```python
#Con el data frame cargado procedemos a aplicar algunos ejercicios de data wrangling
#Filtrar columnas relevantes
felicidad_filtrada = felicidad[['Country name', 'Log GDP per capita', 'Life Ladder', 'Healthy life expectancy at birth','Perceptions of corruption']]
felicidad_filtrada
```

```python
#Para el análisis nos centraremos en los paises de la Alianza del Pacifico
# Lista de países de interés
alianza_pacifico = ['Colombia', 'Chile', 'Mexico', 'Peru']

# Filtrar el DataFrame por los países de interés
felicidad_pacifico = felicidad_filtrada[felicidad_filtrada['Country name'].isin(alianza_pacifico)]
felicidad_pacifico
```

### Uso de loops y funciones

2) Utilizar al menos una vez listas, loops y functions para transformar la base de datos cruda y obtener información relevanete

```python
#Ahora aplicamos loop, listas y funciones
# Función para calcular el promedio de una columna para un país específico
def calcular_promedio_por_pais(dataframe, pais, columna):
    return dataframe[dataframe['Country name'] == pais][columna].mean()

# Listas para almacenar los promedios de esperanza de vida y percepción de corrupción por país
promedios_esperanza_vida = []
promedios_corrupcion = []

# Calcular los promedios para cada país
for pais in alianza_pacifico:
    promedio_esperanza_vida = calcular_promedio_por_pais(felicidad_filtrada, pais, 'Healthy life expectancy at birth')
    promedio_corrupcion = calcular_promedio_por_pais(felicidad_filtrada, pais, 'Perceptions of corruption')

    promedios_esperanza_vida.append((pais, promedio_esperanza_vida))
    promedios_corrupcion.append((pais, promedio_corrupcion))

# Encontrar el país con el promedio de esperanza de vida más alto
pais_max_esperanza_vida = max(promedios_esperanza_vida, key=lambda x: x[1])[0]

# Encontrar el país con la percepción de corrupción promedio más baja
pais_min_corrupcion = min(promedios_corrupcion, key=lambda x: x[1])[0]
```

```python
# Mostrar los resultados
print(f"País con el promedio de esperanza de vida más alto: {pais_max_esperanza_vida}")
print(f"País con la percepción de corrupción promedio más baja: {pais_min_corrupcion}")
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras para el Índice de Felicidad
plt.figure(figsize=(10, 6))
plt.bar(felicidad_pacifico['Country name'], felicidad_pacifico['Life Ladder'], color='skyblue')
plt.xlabel('Países')
plt.ylabel('Índice de Felicidad')
plt.title('Índice de Felicidad por País')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión para Life Ladder y PBI per capitaCorrupción vs. Esperanza de Vida
plt.figure(figsize=(10, 6))
plt.scatter(felicidad_pacifico['Log GDP per capita'], felicidad_pacifico['Life Ladder'])
plt.xlabel('PBI per cápita (log)')
plt.ylabel('Bienestar autopercibido')
plt.title('Bienestar vs. PBI pc')
plt.tight_layout()
plt.show()
```

```python
#Procedemos a colocar las tablas
# Tabla de promedios de variables por país
tabla_promedios = felicidad_pacifico.groupby('Country name').mean().reset_index()
print(tabla_promedios[['Country name', 'Life Ladder', 'Log GDP per capita', 'Healthy life expectancy at birth', 'Perceptions of corruption']])
```

```python
# Finalmente buscamos establecer correlaciones
tabla_correlaciones = felicidad_pacifico[['Life Ladder', 'Log GDP per capita', 'Healthy life expectancy at birth', 'Perceptions of corruption']].corr()
print(tabla_correlaciones)
```

```python
#Agregamos el calculo de una regresión
import statsmodels.api as sm

# Filtrar el DataFrame para las variables de interés
df_regresion = felicidad_pacifico[['Life Ladder', 'Log GDP per capita']]

# Añadir constante para el término independiente (intercepto)
df_regresion = sm.add_constant(df_regresion)

# Definir las variables dependiente e independiente
endogena = df_regresion['Life Ladder']
exogena = df_regresion[['const', 'Log GDP per capita']]

# Crear el modelo de regresión lineal
modelo_regresion = sm.OLS(endogena, exogena)

# Ajustar el modelo a los datos
resultado_regresion = modelo_regresion.fit()

# Imprimir los resultados
print(resultado_regresion.summary())
```

## Ejercicio con NLP

```python
import os
import fitz  
import matplotlib.pyplot as plt

#Ruta de acceso a la carpeta donde están los reportes 2020-2024. Aquí descargamos los reportes en pdf desde enero del 2020 hasta enero del 2024.
ruta_carpeta = "C:\\Users\\ravir\\OneDrive\\Documentos\\Git hub\\Reportes"

#Diccionario para almacenar las frases y sus respectivos valores
frases_dict = {
    "hawkish": ["normalización de la posición de política monetaria", "por encima del rango meta", "retorno de la inflación al rango meta", "normalización de la política monetaria", "atento a la nueva información referida a la inflación y sus determinantes", "compromiso de adoptar las acciones necesarias"],
    "dovish": ["actividad económica en el país está siendo afectada", "reducción de la demanda interna", "sostener el sistema de pagos", "defender la estabilidad macrofinanciera", "debilitamiento significativo de la demanda interna", "actividad económica se encuentra severamente afectada", "riesgos respecto a la actividad económica mundial se han materializado", "recesión global", "actividad económica viene siendo severamente afectada", "debajo del rango meta", "postura monetaria fuertemente expansiva", "política monetaria expansiva" "ampliar el estímulo monetario", "significativamente por debajo de sus niveles del año anterior", "niveles inferiores a los del año pasado", "limite inferior del rango meta", "tramo inferior del rango meta", "debajo de su nivel potencial", "postura monetaria expansiva"],
    "neutro": ["riesgos respecto a la actividad económica mundial por las tensiones comerciales se han atenuado", "atento a la nueva información sobre inflación", "atento a la nueva información sobre las expectativas de inflación", "ciclo de alzas sucesivas", "política monetaria continúa siendo expansiva", "retiro gradual del estímulo monetario", "fin del ciclo de alzas", "condicionados a la nueva información sobre la inflación", "ciclo de sucesivas reducciones"]
}

#Diccionario para almacenar el tono numérico de cada reporte
tono_reportes = {}


#Inicializar contadores de tonos
hawkish_count_total = 0
dovish_count_total = 0
neutro_count_total = 0

#Con este loop procesamos archivo por archivo. Todos están en formato pdf. Este loop trabaja con el texto directo. No tokenizamos el texto dado que en la tokenización se perdían varios aspectos valiosos y no permitía que los datos sean bien interpretados de acuerdo a las keywords establecidas anteriormente.
for archivo in os.listdir(ruta_carpeta):
    if archivo.endswith(".pdf"):
        ruta_archivo = os.path.join(ruta_carpeta, archivo)

        #Leer el contenido del archivo PDF
        with fitz.open(ruta_archivo) as pdf:
            contenido = ""
            for pagina in pdf:
                contenido += pagina.get_text()

        #Inicializar contadores de tonos. Por cada expresión que encuentre en los reportes le sumara 1 a la lista de cada tono, de este modo los reportes
                #que almacenen más palabras de cada tono serán más "fuertemente" hawkish, dovish o neutral. 
        hawkish_count = 0
        dovish_count = 0
        neutro_count = 0

        #Verificar el tono de cada frase.
        for tono, frases in frases_dict.items():
            for frase in frases:
                if frase.lower() in contenido.lower():
                    if tono == "hawkish":
                        hawkish_count += 1
                        hawkish_count_total +=1
                    elif tono == "dovish":
                        dovish_count += 1
                        dovish_count_total +=1
                    elif tono == "neutro":
                        neutro_count += 1
                        neutro_count_total += 1

        #Determinar el tono numérico. Lo restamos por si es que aparece ambos tipo de frases en cada reporte. De modo que, sepamos el grado de tono de cada uno.
        tono_numerico = hawkish_count - dovish_count

        #Almacenar el tono en el diccionario
        tono_reportes[archivo] = tono_numerico


#No es necesario hacer un código para ordenar los reportes pues el nombre de cada pdf hace que se orden de menor a mayor, lo cual es, desde enero del 2020 hasta enero del 2024.
tono_reportes


#Crear un gráfico de barras
categorias = ["Hawkish", "Dovish", "Neutro"]
counts = [hawkish_count_total, dovish_count_total, neutro_count_total]

plt.bar(categorias, counts, color=['red', 'blue', 'green'])
plt.title('Número total de palabras por categoría')
plt.xlabel('Categoría')
plt.ylabel('Número total de palabras')
plt.show()
```

Este gráfico de barras nos dice que en el periodo elegido (enero 2020-enero 2024) vemos que hay más frases/palabras que están categorizadas, según nuestro diccionario, como dovish. Lo cual está relacionado con que los periodos elegidos son sobretodo en el tiempo de la pandemia, donde los bancos centrales del mundo, entre ellos el BCRP, llevó a cabo una política monetaria expansiva.

```python
import matplotlib.pyplot as plt
import pandas as pd

#Convertir el diccionario en un DataFrame
df = pd.DataFrame(list(tono_reportes.items()), columns=['Archivo', 'Tono'])

#Extraer año y mes del archivo y crear las columnas correspondientes
df[['Año', 'Mes']] = df['Archivo'].str.extract(r'-(\d{4})-(\d{2})\.pdf')

#Convertir las columnas de año y mes a números enteros
df['Año'] = df['Año'].astype(int)
df['Mes'] = df['Mes'].astype(int)

#Ordenar DataFrame por año y mes
df = df.sort_values(by=['Año', 'Mes']).reset_index(drop=True)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Crear un gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(df['Año'].astype(str) + '-' + df['Mes'].astype(str), df['Tono'], marker='o', label='Valores mensuales')
plt.axhline(0, color='black', linestyle='--')

plt.title('Orientación de la política monetaria de acuerdo a los programas monetarios del BCRP (2020-2024)')
plt.xlabel('Fecha')
plt.ylabel('Tono de política monetaria')

#Establecer etiquetas del eje x cada 6 meses
inicios_de_año = df[df['Mes'] == 1]['Año'].astype(str) + '-01'
indices_cada_6_meses = np.arange(0, len(df), 6)
etiquetas = [str(df['Año'].iloc[i]) + '-' + str(df['Mes'].iloc[i]).zfill(2) for i in indices_cada_6_meses]
plt.xticks(indices_cada_6_meses, labels=etiquetas, rotation=45, ha='right')

#Agregar cuadro de leyenda en la esquina inferior derecha
legend_labels = {'Hawkish > 0': 'black', 'Dovish < 0': 'black', 'Neutral = 0': 'black'}
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for color in legend_labels.values()]
plt.legend(handles, legend_labels.keys(), loc='lower right', title='Interpretación', bbox_to_anchor=(1, 0), borderaxespad=0.)

plt.tight_layout()
plt.show()
```

En el ejercicio número 2 elegimos trabajar con las notas informativas del programa monetario del Banco Central de Reserva del Perú (BCRP) para obtener por medio de NLP el tono de la política monetaria implementada desde inicios del 2020 hasta el último reporte disponible, es decir, enero del 2024. Los tipos/estilos/tonos de política monetaria pueden ser divididos en tres tipos: dovish, hawkish o neutral. La primera hace referencia a un contexto en que la política monetaria se muestra más adversa al desempleo, y prioriza, en cierta medida, la actividad económica. Por el contrario, una política monetaria más hawkish es adversa a la inflación, por lo cual tratará de atacarla rápidamente, a costa de afectar la actividad económica y el empleo. Por último, una política monetaria neutral es aquella que muestra un caracter incierto, en el sentido de que podría tornarse en cierto momento más hawkish o más dovish, por lo general, se caracterizará por estar relacionados a fragmentos de los reportes donde se señala que los policymakers están atentos a obtener más información para aclarar mejor su visión sobre la economía.

Es relevante este análisis pues permite apreciar la importancia de la comunicación en la tarea de la política monetaria. De acuerdo a palabras clave, logramos obtener, en cierta forma, un indicador de cual es la orientación de la política monetaria. De esta forma, la comunicación de política se vuelve un arma más para los hacedores de política para lograr sus objetivos.

En el gráfico podemos ver que tras empezar la pandemia el BCRP inicia un periodo de casi 2 años en los que lleva a cabo una política que priorizó la actividad económica, es decir, dovish. No es sino hasta inicios del 2022 que el BCRP de manera más clara cambia la orientación de su política monetaria hacia una más hawkish. Esta información obtenida a partir de palabras de los reportes de política monetaria siguen una tendencia similar a la de la tasa de interés de referencia, pues en los años inmediato posteriores a la pandemia la tasa de interés se mantuvo en un nivel bajo, hasta que a finales del 2021 e inicios del 2022 se inició una serie de alzas para combatir la inflación global.
